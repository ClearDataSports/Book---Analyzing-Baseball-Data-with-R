Scatterplots are what you will probably use the most to look at value distribution.

```r
# Data import and set up.  See 3.Intro for more detailed information.
hof <- read.csv("data/hofbatting.csv")
hof$MidCareer <- with(hof, (From + To) / 2)
hof$era <- cut(hof$MidCareer, breaks=c(1800,1900,1919,1941,1960,1976,1993,2999), labels=c("19th Century","Lively Ball","Dead Ball","Integration","Expansion","Free Agency","Long Ball"))
```

```r
# New data feature: Home Run Rate
hof$HR.Rate <- with(hof, HR/AB)
```

```r
# Change the plot size that is drawn so everything will fit.
par(plt=c(.2, .94, .145, .883))

# StripChart showing concentration of values.
#  las attribute is for Horizontal labels.  
stripchart(HR.Rate ~ era, data=hof, method="jitter", pch=1, las=2)
```

Next are box plots.  Also know as bullet graphs.  One of my favorite graphical tools to use, right behind candlesticks.

There are 4 elements to a BP.
1. Lower Quartile.
2. Upper Quartile.
3. Median
4. Lines out to the extreme values.


```r
# Make the draw area bigger so that left labels won't get cut off.
par(plt=c(.2, .94, .145, .883))

# Draw the Box Plot
boxplot(HR.Rate ~ era, data=hof, las=2, horizontal=TRUE, xlab="HR Rate")
```
