```r
# Data import and set up.  See 3.Intro for more detailed information.
hof <- read.csv("data/hofbatting.csv")
hof$MidCareer <- with(hof, (From + To) / 2)
hof$era <- cut(hof$MidCareer, breaks=c(1800,1900,1919,1941,1960,1976,1993,2999), labels=c("19th Century","Lively Ball","Dead Ball","Integration","Expansion","Free Agency","Long Ball"))

```

```r
# Save a bar graph
png("bargraph.png")
barplot(table(hof$era), xlab="Era", ylab="Frequency", main="Era of the non-pitching Hall of Famers")
# dev.off() will actually render the file.
dev.off()
```
