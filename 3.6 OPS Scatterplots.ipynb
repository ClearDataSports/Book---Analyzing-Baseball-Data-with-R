Scatterplots are what you will probably use the most to look at value distribution.

```r
# Data import and set up.  See 3.Intro for more detailed information.
hof <- read.csv("data/hofbatting.csv")
hof$MidCareer <- with(hof, (From + To) / 2)
```

```r
# Basic Scatterplot with labels
plot(hof$MidCareer, hof$OPS, xlab="Mid Career", ylab="OPS", main="OPS Mid Career Year")

# Now add lowess line.  F property is the degree of smoothness between 0-1.  Higher is smoother.
#  lowess is a locally-Weighted Polynomial Regression
lines(lowess(hof$MidCareer, hof$OPS, f=0.3), col="blue")

# Add Regression Line
abline(lm(hof$OPS~hof$MidCareer), col="red")

# Identify doesn't seem to work in Jupyter, should provide clickable names for plot points.
identify(hof$MidCareer, hof$OPS, hof$X, n=4)
```
