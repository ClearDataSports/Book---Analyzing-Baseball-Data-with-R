Important stuff here starting in 3.5.  Introduced is the concept of distribution.  How the values are distributed; are they symmetric or left/right skewed?  

Charts provide a quick visual way to look at the value distribution.


```r
# Data import and set up.  See 3.Intro for more detailed information.
hof <- read.csv("data/hofbatting.csv")
hof$MidCareer <- with(hof, (From + To) / 2)
```

```r
# StripChart to show how the data is distributed throughout the years
stripchart(hof$MidCareer, method="jitter", pch=1, xlab="Mid Career")
```

```r
# Histogram to show how the data is distributed throughout the years 
hist(hof$MidCareer, xlab="Mid Career", main="")
```

```r
# Histogram to show how the data is distributed throughout the years.  Use custom breaks to group by every 20 years.
hist(hof$MidCareer, xlab="Mid Career", main="", breaks=seq(1880, 2000, by=20))
```
