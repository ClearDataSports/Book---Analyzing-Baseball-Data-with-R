Scatterplots are what you will probably use the most to look at value distribution.

```r
# Data import and set up.  See 3.Intro for more detailed information.
hof <- read.csv("data/hofbatting.csv")
```

```r
# Basic Scatterplot with labels
with(hof, plot(OBP, SLG, xlab="OBP", ylab="SLG", main="OBP/SLG"))

# Now remove the outliers that we see in the first plot
# PCH property is for the plotting symbol to use, in this case a small black circle.
with(hof, plot(OBP, SLG, xlim=c(0.25, 0.5), ylim=c(0.28, .75), pch=19, xlab="OBP", ylab="SLG", main="OBP/SLG Limited"))
```

```r
# Same scatterplot as above.
with(hof, plot(OBP, SLG, xlim=c(0.25, 0.5), ylim=c(0.28, .75), pch=19, xlab="OBP", ylab="SLG", main="OBP/SLG Limited"))

# Now add some indicators for OPS.  
# Curve function is really useful as you can use a function as the argument, although we aren't doing that here.
# More info on Curve: http://thomasleeper.com/Rcourse/Tutorials/curve.html
curve(.7-x, add=TRUE)
curve(.8-x, add=TRUE)
curve(.9-x, add=TRUE)
curve(1.0-x, add=TRUE)

# Now add labels so we know what these lines mean.  X,Y coordinates for where to draw the text on the chart along with label text.
text(.27, .44, "OPS 0.7")
text(.27, .54, "OPS 0.8")
text(.27, .64, "OPS 0.9")
text(.27, .74, "OPS 1.0")

# Identity labels for those running in RStudio.
with(hof, identify(OBP, SLG, X, n=6))
```
