```r
# Data import and set up.  See 3.Intro for more detailed information.
hof <- read.csv("data/hofbatting.csv")
hof$MidCareer <- with(hof, (From + To) / 2)
hof$era <- cut(hof$MidCareer, breaks=c(1800,1900,1919,1941,1960,1976,1993,2999), labels=c("19th Century","Lively Ball","Dead Ball","Integration","Expansion","Free Agency","Long Ball"))

```

```r
# Create a table from Era data to plot.
t.era <- table(hof$era)
summary(t.era)
```

```r
dotchart(as.numeric(t.era), labels=names(t.era), xlab="Frequency")
```

```r
# Generate a data frame that contains a subset of data that consists of batters who hit 500 or more home runs.
hof.500 <- subset(hof, HR >= 500)

# Order by OPS so that it looks good in the Plot
hof.500 <- hof.500[order(hof.500$OPS),]

# Plot this thing. (yes, "X" is the name of the Player Name feature.)
dotchart(hof.500$OPS, labels=hof.500$X, xlab="OPS")
```
