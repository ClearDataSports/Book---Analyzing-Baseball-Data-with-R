Wanted to take the time to show you what else you can do to view the shape of your data.

```r
# Data import and set up.  See 3.Intro for more detailed information.
hof <- read.csv("data/hofbatting.csv")
```

```r
# Quick way to look at the shape of whatever features you want.
pairs(~SLG+OBP+ASG+WAR.pos+PA,
      data=hof, 
   main="Pairs")

# More/Other Features
pairs(~HR+RBI+SB+CS+BB+SO+BA+OPS,
      data=hof, 
   main="Pairs")

```

```r
# Take a look at the data structure.  We'll want to remove features that aren't numeric.
str(hof)

# Remove the X and RK features, as we can't run a correlation on Player Names, etc.
hof_allNums <- subset(hof, select = -c(X, Rk))

# Look at the structure of the new data frame
str(hof_allNums)
```

```r
# Install the corrr package by un-commenting out the line below.
#install.packages('corrr', repos='http://cran.us.r-project.org')

# Now run the correlate function for all columns.
library(corrr)
rslt <- correlate(hof_allNums)
rslt

# Run the correlate function for specific columns
rslt <- correlate(hof_allNums) %>% focus(HR, RBI, BA, OPS)
rslt

```
